```python
# Historical data - clean date
historical['date'] = pd.to_datetime(historical['Timestamp IST'], format="%d-%m-%Y %H:%M").dt.date

# Sentiment data - clean date
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date

#Check result
print(historical[['Timestamp IST']].head())
print(sentiment[['classification', 'date']].head)
```

```python
# Convert numeric columns properly
historical['Size USD'] = pd.to_numeric(historical['Size USD'], errors='coerce')
historical['Closed PnL'] = pd.to_numeric(historical['Closed PnL'], errors='coerce')
historical['Fee'] = pd.to_numeric(historical['Fee'], errors='coerce')

# Aggregate by date
daily_trades = historical.groupby('date').agg({
    'Account': 'count',          # number of trades
    'Size USD': 'sum',           # total volume
    'Closed PnL': 'sum',         # total profit/loss
    'Fee': 'sum',                # total fees
    'Execution Price': 'mean'    # avg price traded
}).reset_index()

# Rename columns for clarity
daily_trades.rename(columns={
    'Account': 'trade_count',
    'Size USD': 'total_volume_usd',
    'Closed PnL': 'total_pnl',
    'Fee': 'total_fee',
    'Execution Price': 'avg_price'
}, inplace=True)

print(daily_trades.head())
```

```python
# Merge daily_trades with sentiment on date
merged_data = pd.merge(daily_trades, sentiment[['date', 'classification']],
                       on='date', how='inner')

print(merged_data.head(10))
```

```python
sentiment_summary = merged_data.groupby("classification").agg({"trade_count": "mean", "total_volume_usd": "mean", "total_pnl": "mean", "total_fee": "mean"}).reset_index()
print(sentiment_summary)
```

```python
# merged_data CSV में save करना
merged_data.to_csv("merged_data.csv", index=False)

# Colab से download link बनाना
from google.colab import files
files.download("merged_data.csv")
```

# Exploratory Data Analysis

We analyzed trading data with respect to the Fear & Greed Index.

## Key Insights
- Trading activity is the highest during Extreme Fear, showing that panic leads to more trades but with unstable outcomes.
- Neutral sentiment reflects balanced behavior with moderate trades and profits.
- Greed and Extreme Greed phases show fewer trades, but higher profits.

## Conclusion
Market sentiment strongly affects both trading volume and profitability.
- Fear = more trades, unstable profit
- Greed = fewer trades, better profit

## Recommendations
- Apply strict risk management during Fear and Extreme Fear.
- Focus on quality trades during Greed and Extreme Greed.
- Monitor the Fear & Greed Index regularly.
